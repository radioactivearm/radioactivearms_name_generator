```python
import pandas as pd
import numpy as np
```

```python
#file_paths
ss_file_path = "data/Names for Generator - Scottish Surnames.csv"
sf_file_path = "data/Names for Generator - Scottish Fornames.csv"
uss_file_path = "data/Names for Generator - US Census Surnames.csv"
usf_file_path = "data/Names for Generator - US Census Forenames.csv"
```

```python
scottish_surnames = pd.read_csv(ss_file_path)

scottish_forenames = pd.read_csv(sf_file_path)

us_surnames = pd.read_csv(uss_file_path)

us_forenames =pd.read_csv(usf_file_path)
```

Reworking Scottish Surnames


```python
scottish_surnames.drop(columns=["Year", "Rank", "Number"], inplace=True)
scottish_surnames.head()
```

Reworking US surnames

```python
us_surnames.drop(columns=["RANK"], inplace=True)
us_surnames.columns = ["Surname"]
us_surnames["Surname"] = us_surnames["Surname"].str.capitalize()
us_surnames.head()
```

Combining DataFrames dropping common surnames

```python
surnames = pd.concat([us_surnames, scottish_surnames]).drop_duplicates().reset_index(drop=True)
surnames.head()
```

```python
#checking if it worked
did_it_work = (len(scottish_surnames) + len(us_surnames)) != len(surnames)
print(did_it_work)
```

Reworking Scottish Forenames

```python
scottish_forenames.head()
```

```python
scottish_forenames.drop(columns=["Number in 2025", "Rank in 2025", "Rank in 2024", "Rank in 2010s", "Rank in 2000s", "Rank in 1990s", "Rank in 1980s", "Rank in 1970s", "Change in Ranking 2024-2025"], inplace=True)
scottish_forenames.head()
```

```python
# splitting by gender
scottish_forenames_boy = scottish_forenames[scottish_forenames["Sex"] == "Boy"]
scottish_forenames_girl = scottish_forenames[scottish_forenames["Sex"] == "Girl"]
```

Reworking US Forenames

```python
us_forenames.head()
```

```python
us_forenames.drop(columns=["FREQUENCY (COUNT)", "PROPORTION PER 100,000 POPULATION", "CUMULATIVE PROPORTION"], inplace=True)
us_forenames.head()
```

```python
# changing to float64 type
us_forenames["MALE"] = pd.to_numeric(us_forenames["MALE"].astype(str).str.replace(",", ""))
us_forenames["FEMALE"] = pd.to_numeric(us_forenames["FEMALE"].astype(str).str.replace(",", ""))
```

```python
# adding boy/girl column
us_forenames['Sex'] = np.where(us_forenames["MALE"] >= us_forenames["FEMALE"], "Boy", "Girl")
us_forenames.head()
```

```python
us_forenames.drop(columns=["MALE", "FEMALE", "RANK"], inplace=True)
```

```python
us_forenames.columns = ["Name", "Sex"]
us_forenames.head()
```

```python
us_forenames["Name"] = us_forenames["Name"].str.capitalize()
us_forenames.head()
```

```python
us_forenames = us_forenames[["Sex", "Name"]]
```

```python
us_forenames_boy = us_forenames[us_forenames["Sex"] == "Boy"]
us_forenames_girl = us_forenames[us_forenames["Sex"] == "Girl"]
```

```python
us_forenames_boy.head()
```

```python
forenames_boy = pd.concat([us_forenames_boy, scottish_forenames_boy]).drop_duplicates().reset_index(drop=True)
forenames_girl = pd.concat([us_forenames_girl, scottish_forenames_girl]).drop_duplicates().reset_index(drop=True)
forenames_boy.head()
```

```python
did_it_work = (len(scottish_forenames_girl) + len(us_forenames_girl)) != len(forenames_girl)
print(did_it_work)
```

```python
surnames.to_csv("cleaned_data/surnames.csv")
forenames_boy.to_csv("cleaned_data/forenames_boy.csv")
forenames_girl.to_csv("cleaned_data/forenames_girl.csv")
```
